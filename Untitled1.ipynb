```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.sankey import Sankey

# 1. Dati
np.random.seed(0)
n_prodotti = 50
prodotti = [f"Prodotto {i+1}" for i in range(n_prodotti)]
vendite = np.random.exponential(scale=100, size=n_prodotti).round(2)

df = pd.DataFrame({'Prodotto': prodotti, 'Vendite': vendite})
df = df.sort_values(by='Vendite', ascending=False)
df['% Cumulativo'] = df['Vendite'].cumsum() / df['Vendite'].sum() * 100

# Identifica i prodotti che formano l'80% delle vendite
df['Categoria'] = np.where(df['% Cumulativo'] <= 80, 'Top 20%', 'Altri')

# 2. Grafico Pareto
fig, ax1 = plt.subplots(figsize=(10, 5))
ax1.bar(df['Prodotto'], df['Vendite'], color='skyblue')
ax1.tick_params(axis='x', labelrotation=90)
ax1.set_ylabel("Vendite")

ax2 = ax1.twinx()
ax2.plot(df['Prodotto'], df['% Cumulativo'], color='red', marker='o')
ax2.set_ylabel("% Cumulativo")
ax2.axhline(80, color='green', linestyle='--')
plt.title("Curva di Pareto - Vendite")
plt.tight_layout()
plt.show()

# 3. Grafico a torta
fig, ax = plt.subplots()
quota = df['Categoria'].value_counts()
ax.pie(quota, labels=quota.index, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightgray'])
ax.axis('equal')
plt.title("Distribuzione Prodotti - Top 20% vs Altri")
plt.show()
```
